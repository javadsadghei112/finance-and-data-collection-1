```python
import yfinance as yf
import pandas as pd
import requests
from bs4 import BeautifulSoup

#       Wikipedia
def get_sp500_tickers():
    url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    #    
    table = soup.find('table', {'class': 'wikitable'})
    tickers = []

    for row in table.findAll('tr')[1:]:  #    
        ticker = row.findAll('td')[0].text.strip()  #  
        tickers.append(ticker)

    return tickers


```

```python


def filter_companies_by_market_cap(tickers, min_cap):
    data = []

    for ticker in tickers:
        try:
            company = yf.Ticker(ticker)
            info = company.info
            market_cap = info.get("marketCap")

            if market_cap and  market_cap >= min_cap:
                data.append({
                    "Company Name": info.get("longName"),
                    "Symbol": ticker,
                    "Market Cap": market_cap,
                    "Sector": info.get("sector"),
                    "Country": info.get("country")
                })
        except:
            continue

    return pd.DataFrame(data)


tickers = get_sp500_tickers()

min_market_cap = 151e9
filtered_companies = filter_companies_by_market_cap(tickers, min_market_cap)

#  
print(filtered_companies)
```

```python
df = pd.DataFrame(filtered_companies)
df.head(6)
```

```python
df.sort_values(by="Market Cap", ascending=False, inplace=True)
df
```
